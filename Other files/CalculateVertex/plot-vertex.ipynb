```python
import glob
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
from matplotlib.pyplot import cm 
from scipy.optimize import curve_fit

def f(x,a,b):
    return a*x + b

%matplotlib inline

mpl.rc("savefig", dpi=100)

font = {'family' : 'Times New Roman',
        'weight' : 'normal',
        'size'   : 20}

# mpl.rcParams['text.usetex'] = True
# mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}'] #for \text command

mpl.rc('font', **font)

tau = (1+5**.5) / 2
rho = 2*(tau+1) / (tau+2) / (3-tau)**.5
```

```python
files = glob.glob('./out_vert/vert*npz')

counts = []
freqs = []
g = []
s = []

v_types = ['Q', 'R', 'K', 'M', 'D', 'ST', 'Z', 'T', 'U', 'S', 'J', 'L', 'V', 'W', 'X', 'Y']

for f in files:
    t = np.load(f)
    g.append(0.5 - np.abs(t['g']%1-0.5))
    cur_cts = np.sum(t['cts'],axis=0)
#    cur_cts = np.mean(t['cts'],axis=0)
    tot_cts = sum(cur_cts)
    counts.append(cur_cts)
    freqs.append(cur_cts.astype(float) / tot_cts * 100)
    s.append(t['s'])
    
g = np.array(g)
counts = np.array(counts)
freqs = np.array(freqs)

unique_gam = np.unique(g)

print len(unique_gam)
for i in unique_gam:
    print i,','
```

```python
labels = ['Q', 'R', 'K', 'M', 'D', 'ST', 'Z',
          'X', 'U', 'S', 'J', 'L', 'V','W', 'T', 'Y']

m = len(labels)
for i in range(m):
    print i, labels[i]#, freqs[-1,i]
    
np.savez('vert_freq',gam=unique_gam,freqs=freqs,labels=labels)
```

```python
# use broken axis from
# http://matplotlib.org/examples/pylab_examples/broken_axis.html

fig, (ax, ax2) = plt.subplots(2, 1, figsize=(8,12), sharex=True)

ax.tick_params(axis='both',which='both',direction='out')
ax2.tick_params(axis='both',which='both',direction='out')

m = len(v_types)

# plot vertical lines

def EquivGam(g):
    if g % 1 <= 0.5:
        return g % 1
    else:
        return 1 - g % 1

for i in [1,4,2]:
    ax.axvline(EquivGam(i * tau), 0, 1, dashes=[i*2,i,3,i],alpha=0.5,lw=1,color='black')
    ax2.axvline(EquivGam(i * tau), 0, 1, dashes=[i*2,i,3,i],alpha=0.5,lw=1,color='black')

    if i==1:
        text = r'$\tau$'.format(int(i*tau))
    else:
        text = r'{:d}$\tau$'.format(i,int(i*tau))

    x = EquivGam(i*tau)

    ax.annotate(text, family='monospace',
                 #xy=(x,y-0.0025),
                 xy=(x,35),
                xycoords='data',
#                xytext=(0, 320*(-0.45 + EquivGam(i * tau))), textcoords='offset points',
                xytext=(0,0), textcoords='offset points',
                horizontalalignment='center', verticalalignment='center',fontsize=20,
                bbox=dict(facecolor='white', edgecolor='None', alpha=1.0,pad=3.0)
                )



color=cm.jet(np.linspace(0,1,m))


# plot the same data on both axes
for i in range(m):
    ax.plot(unique_gam, freqs[:,i], lw=3, color=color[i])#,marker='o',markersize=1,markerfacecolor=color[i],markeredgecolor=color[i])
    ax2.plot(unique_gam, freqs[:,i], lw=3, color=color[i])#,marker='o',markersize=1,markerfacecolor=color[i],markeredgecolor=color[i])
    
# zoom-in / limit the view to different portions of the data
ax.set_ylim(4.1, 43)  # outliers only
ax2.set_ylim(0, 4.1)  # most of the data

ax.set_yticks(range(10,43,10))
ax2.set_yticks(range(0,5))

# hide the spines between ax and ax2
ax.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop='off')  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

# This looks pretty good, and was fairly painless, but you can get that
# cut-out diagonal lines look with just a bit more work. The important
# thing to know here is that in axes coordinates, which are always
# between 0-1, spine endpoints are at these locations (0,0), (0,1),
# (1,0), and (1,1).  Thus, we just need to put the diagonals in the
# appropriate corners of each of our axes, and so long as we use the
# right transform and disable clipping.

d = .015  # how big to make the diagonal lines in axes coordinates
# arguments to pass plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

# What's cool about this is that now if we vary the distance between
# ax and ax2 via f.subplots_adjust(hspace=...) or plt.subplot_tool(),
# the diagonal lines will move accordingly, and stay right at the tips
# of the spines they are 'breaking'
fig.tight_layout()
plt.xlim([0,0.5])


##


# offset
curind = 2
r = np.zeros((m,2))
print labels[curind], freqs[-1,curind]*100


init_ax = [ax for x in range(m)]

init_r = np.zeros((m,2))

visible = np.ones(m)
visible[5] = 0 #ST
visible[6] = 0 #Z
visible[9] = 0
visible[15] = 0

init_ax[0] = ax
init_ax[1] = ax2
init_ax[2] = ax
init_ax[3] = ax2
init_ax[4] = ax
init_ax[5] = ax2
init_ax[6] = ax2
init_ax[7] = ax2
init_ax[8] = ax2
init_ax[9] = ax
init_ax[10] = ax
init_ax[11] = ax2
init_ax[12] = ax2
init_ax[13] = ax2
init_ax[14] = ax2
init_ax[15] = ax2

"""
0 Q
1 R
2 K
3 M
4 D
5 ST
6 Z
7 X
8 U
9 S
10 J
11 L
12 V
13 W
14 T
15 Y
"""
r[0] = [0.03, -0.5] #Q
r[1] = [0.03,0] #R
r[2] = [0.03,-3] #K
r[3] = [0.03,2] #M
r[4] = [0.03,5] #D
r[5] = [0.03,-0.2] #ST
r[6] = [0.03,-0.2] #Z
r[7] = [0.03,-0.3] #X
r[8] = [0.03,0] #U
r[9] = [0.03,0] #S
r[10] = [0.03,0] #J
r[11] = [0.03,0] #L
r[12] = [0.03,0] #V
r[13] = [0.03,.02] #W
r[14] = [0.03,0] #T
r[15] = [0.03,-0.35] #Y

init_r[0] = [0.03, 16.25]
init_r[1] = [0.22, 1.]
init_r[2] = [0.03, 10.5]
init_r[3] = [0.07, 3.7]
init_r[4] = [0.03, 38]
init_r[7] = [0.17, 1.3]
init_r[8] = [0.25, 0.72]
init_r[9] = [0.03,7]
init_r[10] = [0.03, 23]
init_r[11] = [0.036, 3.9]
init_r[12] = [0.19, 3.8]
init_r[13] = [0.275, 0.31]
init_r[14] = [0.12, 3.3]

fsize = 20
for i in range(m):
    x = 0.5
    y = freqs[-1,i]
    if visible[i]:
        init_ax[i].annotate(labels[i], family='monospace',xy=(init_r[i,0],init_r[i,1]),
                xycoords='data',
                xytext=(0,0),
                textcoords='offset points',
                horizontalalignment='center', verticalalignment='center',fontsize=fsize)
    if i == 7:
        #extra label for X
        cur_ax.annotate(labels[i], family='monospace',xy=(0.35,2),
                xycoords='data',
                xytext=(0.35,2),
                textcoords='data',
                horizontalalignment='left', verticalalignment='center',fontsize=fsize)
    if i == 9:
        init_ax[i].annotate('S', family='monospace',xy=(init_r[i,0],init_r[i,1]),
                xycoords='data',
                xytext=(0,0),
                textcoords='offset points',
                horizontalalignment='center', verticalalignment='center',fontsize=fsize)
        init_ax[i].annotate('V', family='monospace',xy=(init_r[i,0],init_r[i,1]),
                xycoords='data',
                xytext=(-4,-30),
                textcoords='offset points',
                horizontalalignment='center', verticalalignment='center',fontsize=fsize)



    if np.isnan(y):
        print labels[i]
        continue
    else:
        if y > 5:
            cur_ax = ax
        else:
            cur_ax = ax2
        if freqs[-1,i] < 10**-5:
            continue
        cur_ax.annotate(labels[i], family='monospace',xy=(x,y),
                xycoords='data',
                xytext=(x+r[i,0], y + r[i,1]),
                textcoords='data',
                arrowprops=dict(color=color[i],headwidth=5,frac=0.3,shrink=0.2,width=1),
                horizontalalignment='left', verticalalignment='center',fontsize=fsize)

ax.set_ylabel(r'$F$  ' + r"[%]",fontsize=25)
plt.xlabel(r'$\gamma$',fontsize=25)
#fig.tight_layout()
fig.subplots_adjust(left=0.1,right=0.9,bottom=0.07)
fig.savefig('fig_vertex_freq.pdf',dpi=300)
```
