```python
import glob
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
from matplotlib.pyplot import cm 
from scipy.optimize import curve_fit

def f(x,a,b):
    return a*x + b

%matplotlib inline

mpl.rc("savefig", dpi=100)

font = {'family' : 'Times New Roman',
        'weight' : 'normal',
        'size'   : 15}

# mpl.rcParams['text.usetex'] = True
# mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}'] #for \text command

mpl.rc('font', **font)

tau = (1+5**.5) / 2
rho = 2*(tau+1) / (tau+2) / (3-tau)**.5
```

```python
files = glob.glob('./vor2/vor2*k_100*npz')

counts = []
g = []
s = []
areas = []
freqs = []

for f in files:
    t = np.load(f)
    g.append(0.5 - np.abs(t['g']%1-0.5))
    counts.append(t['cts'])
    freqs.append(t['freq'])
    areas.append(t['areas'])
    s.append(t['s'])
    
g = np.array(g)
areas = np.array(areas)#[ind]
freq = np.array(freqs)#[ind]
counts = np.array(counts)

unique_gam = np.unique(g)

print len(unique_gam)
for i in unique_gam:
    print i,','
    
all_areas = np.concatenate(areas)
unique_areas = np.unique(np.round(all_areas,5))
```

```python
n = len(unique_gam)
m = len(unique_areas)

freqs = []
err = []
for i in range(n):
    ind = np.where(g == unique_gam[i])[0]
    cur_counts_g = []
    for j in ind:
        cur_counts_s = np.zeros(m)
        cur_areas = np.round(areas[j],5)
        for k in range(len(cur_areas)):
            a = cur_areas[k]
            area_ind = np.where(unique_areas == a)[0][0]
            cur_counts_s[area_ind] += counts[j][k]
        cur_counts_g.append(cur_counts_s)
    cur_freqs_g = np.einsum('ij,i->ij',cur_counts_g, 1/ np.sum(cur_counts_g,axis=1))
    std_freqs_g = np.std(cur_freqs_g,axis=0)
    avg_freqs_g = np.sum(cur_counts_g,axis=0) / np.sum(np.sum(cur_counts_g,axis=0))
    freqs.append(avg_freqs_g)
    err.append(std_freqs_g)

err = np.array(err)
freqs = np.array(freqs)

def zero_to_nan(values):
    """Replace every 0 with 'nan' and return a copy."""
    return [float('nan') if x<10**-10 else x for x in values]

freqs_nozero = [zero_to_nan(x) for x in freqs]
freqs_nozero = np.array(freqs_nozero)
```

```python
# use broken axis from
# http://matplotlib.org/examples/pylab_examples/broken_axis.html

fig, (ax, ax2) = plt.subplots(2, 1, figsize=(4,6), sharex=True)

# plot vertical lines

def EquivGam(g):
    if g % 1 <= 0.5:
        return g % 1
    else:
        return 1 - g % 1

for i in range(1,6):
    ax.axvline(EquivGam(i * tau), 0, 1, dashes=[i*2,i,3,i],alpha=0.5,lw=1,color='black')
    ax2.axvline(EquivGam(i * tau), 0, 1, dashes=[i*2,i,3,i],alpha=0.5,lw=1,color='black')


color=cm.jet(np.linspace(0,1,m))

# plot the same data on both axes
for i in range(m):
    ax.plot(unique_gam, freqs[:,i]*100, lw=1, color=color[i])
    ax2.plot(unique_gam, freqs[:,i]*100, lw=1, color=color[i])
    ax.scatter(unique_gam, freqs_nozero[:,i]*100, color=color[i],s=15)
    ax2.scatter(unique_gam, freqs_nozero[:,i]*100, color=color[i],s=15)

# zoom-in / limit the view to different portions of the data
ax.set_ylim(5, 43)  # outliers only
ax2.set_ylim(0, 4)  # most of the data

ax.set_yticks(range(10,43,10))
ax2.set_yticks(range(0,5))

# hide the spines between ax and ax2
ax.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax.xaxis.tick_top()
ax.tick_params(labeltop='off')  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

# This looks pretty good, and was fairly painless, but you can get that
# cut-out diagonal lines look with just a bit more work. The important
# thing to know here is that in axes coordinates, which are always
# between 0-1, spine endpoints are at these locations (0,0), (0,1),
# (1,0), and (1,1).  Thus, we just need to put the diagonals in the
# appropriate corners of each of our axes, and so long as we use the
# right transform and disable clipping.

d = .015  # how big to make the diagonal lines in axes coordinates
# arguments to pass plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal
ax.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # top-right diagonal

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # bottom-right diagonal

# What's cool about this is that now if we vary the distance between
# ax and ax2 via f.subplots_adjust(hspace=...) or plt.subplot_tool(),
# the diagonal lines will move accordingly, and stay right at the tips
# of the spines they are 'breaking'
fig.tight_layout()
plt.xlim([0,0.5])

ax.set_ylabel(r'$P(A_\mathrm{Vor}^2)$' + "[\%]")
plt.xlabel(r'$\sum\gamma_j$')
```

```python
n = len(unique_gam)
mn = []
std = []
cts = []

for j in range(n):
    ind = np.where(g == unique_gam[j])[0]
    cur_mn = 0
    cur_var = 0
    total = sum([sum(x) for x in counts[ind]])
    for i in ind:
        cur_mn += np.sum(areas[i] * counts[i])
    cur_mn /= total
    for i in ind:
        cur_var += np.sum((areas[i]-cur_mn)**2 * counts[i])
    cur_var /= total
    mn.append(cur_mn)
    std.append(cur_var**0.5)
    avg_nbr = np.median([len(x) for x in counts[ind]])
    cts.append(avg_nbr)
    print '{:0.5f} {:0.3f} {:0.5f} {:0.2f}'.format(float(unique_gam[j]), cur_mn, cur_var**0.5, avg_nbr)
    
mn = np.array(mn)
std = np.array(std)
cts = np.array(cts)
```
