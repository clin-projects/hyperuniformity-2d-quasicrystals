```python
import glob
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
from matplotlib.pyplot import cm 
from scipy.optimize import curve_fit

def f(x,a,b,c):
    return a*x# + b# + b*np.log(x) + c

%matplotlib inline

mpl.rc("savefig", dpi=100)

font = {'family' : 'Times New Roman',
        'weight' : 'normal',
        'size'   : 20}

# mpl.rcParams['text.usetex'] = True
# mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}'] #for \text command

mpl.rc('font', **font)

def CalcSmoothN(x,y,n):
    # does cumulative average
    # applies trapezoidal rule, returns new [x,y]
    #
    smoothy = []
    midpt_x = []
    cursum = 0
    N = len(y)
    x0 = (x[0] + x[1])/2
    for i in range(1,N-1):
        cur_integrand = (y[i]*x[i]**n + y[i+1]*x[i+1]**n)/2
        cur_x = (x[i] +x[i+1])/2
        dx = x[i+1] - x[i]
        cursum += cur_integrand * dx
        smoothy.append(cursum / (cur_x - x0)**(n+1))
        midpt_x.append(cur_x)
    return np.array(midpt_x), np.array(smoothy)

def EquivGam(g):
    if g % 1 <= 0.5:
        return g % 1
    else:
        return 1 - g % 1

tau = (1+5**.5) / 2
rho = 2*(tau+1) / (tau+2) / (3-tau)**.5
```

```python
files = glob.glob('./out/*.npz')

m = len(files)

gam = []
kmax = []
seeds = []
nwin = []
for i in range(m):
    t = np.load(files[i])
    nwin.append(int(t['num_windows']))
    gammasum = float(t['gammasum'])
    kmax.append(int(t['kmax']))
    seeds.append(int(t['seed']))
    gam.append(round(gammasum % 1,10))
    
    fsplit = files[i].split('_')
    ind = fsplit.index('g')+1
    cur_g_str = fsplit[ind]
    
    if round(gammasum % 1,10) != float(cur_g_str):
        print i
        fsplit = files[i].split('_')
        ind = fsplit.index('g')+1
        cur_g_str = fsplit[ind]
        new_fname = files[i].replace('g_'+cur_g_str,'g_'+str(round(gammasum % 1, 10)))
        os.rename(files[i], new_fname)

kmax = np.array(kmax)
seeds = np.array(seeds)
gam = np.array(gam)
nwin = np.array(nwin)
unique_gam = np.unique(gam)

print len(unique_gam)
for i in unique_gam:
    print i,','
```

```python
# plot versus R

ind = np.where(nwin == 10000)[0]

unique_kmax = np.unique(kmax[ind])

m = len(unique_kmax)

coeff = []
fits = []
err = []

for j in range(m):
    k = unique_kmax[j]
    print k,
    fits_k = []
    ind_k = np.where(kmax[ind] == k)[0]
    coeff2 = []
    print len(ind_k)
    for i in ind_k:
        t = np.load(files[ind[i]]) # load only max_k
        
        var = t['var']
        rlist = t['rlist']
#        print rlist[-1]
        
#        print t['kmax'], t['num_windows']
        
        normvar = var / (np.pi * rho)**.5 / rlist / 2
        

        y = var / (np.pi * rho)**.5 / 2
        x = rlist
        
        n = -1
        xs, ys = CalcSmoothN(x,y,n)
        ys *= (n+2)
        
        po, pc = curve_fit(f,xs,ys)
        fits_k.append(po[0])
        coeff2.append(po[2])
    coeff.append(np.mean(coeff2))
    fits.append(np.mean(fits_k))
    err.append(np.std(fits_k))
coeff = np.array(coeff)
fits = np.array(fits)
err = np.array(err)
```
